````python
import base64
import os
from google import genai
from google.genai import types


def generate():
    client = genai.Client(
        api_key=os.environ.get("GEMINI_API_KEY"),
    )

    model = "gemini-2.5-pro-preview-03-25"
    contents = [
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(text="""Make an app consisting super admin and seller account and customer account for the purpose of milk dairy , it is for where the customer nearer to the milk seller to order the milk , Customer will show the nearer seller it is for rural areas , and there is a condition for sellers is first they have to top up for the account where the seller getting the order 2rupees will debit and credit into super admin account and there is a condition for customer is if he order the milk the seller is accepts if customer cancel and add a fine to customer is 12rupees and I want frontend and backend for this code and code is in react native make it better ui and ux"""),
            ],
        ),
        types.Content(
            role="model",
            parts=[
                types.Part.from_text(text="""Here's a breakdown of the thought process behind generating the React Native Milk Dairy App code and explanation:

1.  **Understanding the Core Request:** The user wants a React Native app for a milk dairy business targeting rural areas. It needs three user roles: Super Admin, Seller, and Customer. Key features include:
    *   Location-based discovery (customer sees nearby sellers).
    *   Seller Wallet: Sellers need to top up. A fee (₹2) is deducted per accepted order and transferred to the Super Admin.
    *   Customer Order Cancellation Fine: A fine (₹12) is applied if a customer cancels an *accepted* order.
    *   Frontend (React Native) and Backend (Conceptual).
    *   Emphasis on good UI/UX.

2.  **Deconstructing the Requirements:**

    *   **User Roles:**
        *   **Super Admin:** Manages sellers, views transactions, manages wallet system (conceptually), views overall stats. Needs a separate interface (likely web-based, but the request is for React Native, so maybe a simplified RN admin panel).
        *   **Seller:** Registers, manages profile (location is key), manages milk availability/stock, views/accepts/rejects orders, manages wallet (top-up, view balance, transaction history), views order history.
        *   **Customer:** Registers, manages profile (location is key), views nearby sellers, views seller profiles/milk details, places orders, views order history, cancels orders (with potential fine), manages profile.
    *   **Core Functionality:**
        *   **Authentication:** Login/Signup for all roles.
        *   **Location:** Crucial for finding nearby sellers. Needs device location permissions and potentially a mapping library.
        *   **Seller Discovery:** Filtering sellers based on customer's location.
        *   **Ordering:** Creating, viewing, accepting/rejecting, cancelling orders.
        *   **Wallet System:**
            *   Seller Top-up (mechanism needed - integration with payment gateway or manual admin confirmation).
            *   Order Fee Deduction (triggered on order acceptance).
            *   Fine Application (triggered on customer cancellation *after* acceptance).
            *   Transaction History for Seller and Admin.
    *   **Technology:**
        *   **Frontend:** React Native.
        *   **Backend:** Needs to be specified (Node.js/Express, Python/Django/Flask, Firebase, etc.). The request asks for *conceptual* backend, so I'll outline the API endpoints and data models.
    *   **UI/UX:** Clean design, intuitive navigation, clear information display, feedback mechanisms (loading states, success/error messages).

3.  **Structuring the App (React Native):**

    *   **Navigation:** Use React Navigation (`@react-navigation/native`, `stack`, `bottom-tabs`).
        *   Initial Auth flow (Login/Signup).
        *   Separate stacks/tabs for Customer, Seller, and potentially Admin (though a dedicated web admin panel is more common).
    *   **Folder Structure:** Organize components, screens, navigation, services (API calls), state management (if needed), assets, utils.
    *   **Key Screens:**
        *   **Auth:** LoginScreen, SignupScreen.
        *   **Customer:**
            *   HomeScreen (Map/List of nearby sellers).
            *   SellerDetailsScreen (View products, place order).
            *   OrdersScreen (View current/past orders).
            *   ProfileScreen.
            *   OrderConfirmationScreen.
            *   OrderCancellationScreen (with fine warning).
        *   **Seller:**
            *   DashboardScreen (Overview, new orders).
            *   OrdersScreen (View pending/accepted/completed/rejected orders).
            *   WalletScreen (View balance, top-up history, transaction log).
            *   ProfileScreen (Manage details, location, availability).
            *   TopUpScreen (Initiate top-up).
        *   **Admin (Simplified RN version):**
            *   SellerListScreen.
            *   TransactionListScreen.
            *   SettingsScreen (e.g., set fees).
    *   **State Management:** For simpler apps, `useState` and `useContext` might suffice. For more complex state (like user auth, shared data), consider Zustand, Redux Toolkit, or Jotai. I'll start with `useState` and `useContext` for simplicity in the example.
    *   **UI Library:** React Native Elements or React Native Paper can speed up UI development and provide consistent styling. I'll use core React Native components plus potentially React Native Elements for structure.
    *   **Location:** `react-native-geolocation-service` or Expo's Location API. Needs permission handling.
    *   **Mapping:** `react-native-maps` for displaying sellers on a map.

4.  **Designing the Backend (Conceptual):**

    *   **Database Schema (e.g., MongoDB/NoSQL):**
        *   `users`: { `_id`, `name`, `email`, `passwordHash`, `role` ('customer', 'seller', 'admin'), `phone`, `address`, `location` (GeoJSON Point: { `type`: 'Point', `coordinates`: [longitude, latitude] }), `walletBalance` (for seller), `createdAt`, `updatedAt` }
        *   `sellers`: { `_id`, `userId` (ref to users), `shopName`, `availabilityStatus`, `deliveryRadius` (optional), `products` [{ `productId`, `name`, `price`, `unit`, `stock` }], `createdAt`, `updatedAt` } // Could merge some into users if 1:1
        *   `products`: { `_id`, `name`, `description`, `defaultUnit` } // Optional: If products are standardized
        *   `orders`: { `_id`, `customerId` (ref to users), `sellerId` (ref to users), `items` [{ `productId`, `name`, `price`, `quantity`, `unit` }], `totalAmount`, `status` ('pending', 'accepted', 'rejected', 'out_for_delivery', 'delivered', 'cancelled_by_customer', 'cancelled_by_seller'), `orderTimestamp`, `acceptedTimestamp`, `deliveredTimestamp`, `cancellationTimestamp`, `cancellationFineApplied` (boolean), `deliveryAddress`, `location` (customer's location at order time) }
        *   `transactions`: { `_id`, `userId` (ref to users - seller or admin), `type` ('topup', 'order_fee', 'cancellation_fine_credit', 'cancellation_fine_debit', 'withdrawal'), `amount`, `orderId` (ref to orders, optional), `description`, `timestamp` }
    *   **API Endpoints:**
        *   `POST /auth/register`
        *   `POST /auth/login`
        *   `GET /auth/me` (Get current user details)
        *   `PUT /users/profile` (Update profile, location)
        *   `GET /sellers/nearby?lat=...&lon=...&radius=...` (Find sellers near customer)
        *   `GET /sellers/:id` (Get seller details)
        *   `PUT /sellers/profile` (Seller updates profile/availability)
        *   `POST /sellers/wallet/topup` (Initiate top-up - might need admin approval flow)
        *   `GET /sellers/wallet/balance`
        *   `GET /sellers/wallet/transactions`
        *   `POST /orders` (Customer places order)
        *   `GET /orders/customer` (Customer views their orders)
        *   `GET /orders/seller` (Seller views their orders)
        *   `GET /orders/:id` (View specific order details)
        *   `PUT /orders/:id/accept` (Seller accepts) -> Trigger fee deduction
        *   `PUT /orders/:id/reject` (Seller rejects)
        *   `PUT /orders/:id/cancel/customer` (Customer cancels) -> Check if accepted, apply fine
        *   `PUT /orders/:id/status` (Update status like delivered)
        *   **Admin Endpoints:**
            *   `GET /admin/sellers`
            *   `GET /admin/transactions`
            *   `PUT /admin/sellers/:id/approve` (If needed)
            *   `POST /admin/wallet/confirm_topup` (If manual top-up)

5.  **Implementing Key Logic:**

    *   **Nearby Sellers:** Use database geospatial queries (e.g., MongoDB's `$nearSphere`).
    *   **Fee Deduction:** In the `PUT /orders/:id/accept` backend logic, after updating the order status:
        1.  Check seller's wallet balance >= fee (₹2). If not, reject acceptance? Or allow negative balance? (Decision needed). *Let's assume sufficient balance is required.*
        2.  Decrement seller's `walletBalance` by ₹2.
        3.  Increment Super Admin's effective balance (or a central account) by ₹2.
        4.  Create two `transaction` records: one debit for the seller, one credit for the admin.
    *   **Cancellation Fine:** In the `PUT /orders/:id/cancel/customer` backend logic:
        1.  Fetch the order.
        2.  Check if `order.status === 'accepted'`.
        3.  If yes:
            *   Update order status to `cancelled_by_customer`.
            *   Set `cancellationFineApplied = true`.
            *   Create a `transaction` record for the customer (type `cancellation_fine_debit`, amount ₹12). *How is the fine collected? Add to next order? Separate payment? This needs clarification. For now, just log it.*
            *   Create a `transaction` record for the admin (type `cancellation_fine_credit`, amount ₹12). *This assumes the fine goes to the admin.*
        4.  If no (status was 'pending'), just update status to `cancelled_by_customer`.

6.  **Refining UI/UX:**

    *   **Clear Navigation:** Tabs for main sections, clear back buttons in stacks.
    *   **Visual Feedback:** Loading spinners, success/error messages (e.g., using `react-native-toast-message`).
    *   **Map Integration:** Show seller locations clearly on a map. Tapping a marker navigates to the seller's details.
    *   **Order Flow:** Simple steps: View Seller -> Add to Cart (Implicitly, for milk it might be just quantity) -> Confirm Order -> View Status.
    *   **Wallet:** Clear display of balance, easy access to transaction history. Top-up process should be straightforward (even if conceptual).
    *   **Error Handling:** Gracefully handle API errors, location permission denial.
    *   **Empty States:** Show informative messages when lists are empty (no nearby sellers, no orders, no transactions).
    *   **Consistency:** Use a consistent color scheme, typography, and component style.

7.  **Writing the Code (React Native):**

    *   Set up the basic project structure.
    *   Implement navigation.
    *   Create placeholder screens.
    *   Develop Auth screens and logic (using dummy data or mock API calls initially).
    *   Implement Customer flow (location, seller list, order).
    *   Implement Seller flow (order management, wallet).
    *   Add UI elements using React Native core components and potentially a library.
    *   Integrate location services and maps.
    *   Define API service functions (using `fetch` or `axios`).
    *   Add state management where needed.
    *   Style the components.

8.  **Adding Explanations and Caveats:**

    *   Clearly state that the backend is conceptual.
    *   Mention necessary libraries (`react-navigation`, `react-native-maps`, `react-native-geolocation-service`, potentially UI kit, state management).
    *   Highlight assumptions made (e.g., fine collection mechanism, top-up process).
    *   Explain the core logic (fee deduction, fine).
    *   Emphasize the need for a real backend implementation.
    *   Suggest improvements (real-time updates, notifications, payment gateway integration).

**(Self-Correction during thought process):**

*   Initially, I might forget the fine is *only* if the order is *accepted*. Need to add that check.
*   How does the seller top up? Needs clarification. I'll suggest a conceptual API endpoint and mention it might involve payment gateways or manual admin confirmation.
*   How is the customer fine collected? This is a business logic gap. I'll note it and create the transaction log entry, assuming the collection mechanism is handled elsewhere (e.g., added to the next bill, blocked ordering until paid).
*   An Admin RN app might be clunky. It's better as a web app, but the user requested RN. I'll provide a *basic* RN admin structure but recommend web.
*   Need to handle location permissions explicitly.
*   Need loading states and error handling for good UX.
*   Need to define the data structures (interfaces/types) for clarity.

By following these steps, the comprehensive response including conceptual backend, frontend structure, code snippets, and explanations can be generated."""),
                types.Part.from_text(text="""Okay, let's outline the structure and provide conceptual code for this React Native Milk Dairy App.

**Core Concepts & Technologies:**

1.  **Frontend:** React Native (using Expo for easier setup or React Native CLI)
2.  **Navigation:** React Navigation (`@react-navigation/native`, `@react-navigation/stack`, `@react-navigation/bottom-tabs`)
3.  **UI Components:** React Native core components, potentially a UI library like React Native Elements or React Native Paper for faster styling.
4.  **Location:** `expo-location` (if using Expo) or `react-native-geolocation-service` + `react-native-permissions` (for RN CLI).
5.  **Maps:** `react-native-maps` for displaying sellers.
6.  **State Management:** React Context API for managing user auth and role, potentially Zustand or Redux Toolkit for more complex state.
7.  **Backend:** Conceptual (API endpoints and logic described). You'll need to build this separately using Node.js/Express, Python/Django/Flask, Firebase Functions, etc.
8.  **Database:** A database that supports geospatial queries (like MongoDB, PostgreSQL with PostGIS) is crucial for finding nearby sellers.

---

**I. Backend Design (Conceptual)**

You'll need a server application with a database.

**1. Database Models:**

*   **Users:**
    *   `userId` (Primary Key)
    *   `name`
    *   `email` (unique)
    *   `passwordHash`
    *   `phone` (unique)
    *   `role` ('customer', 'seller', 'superadmin')
    *   `address` (text)
    *   `location` (GeoJSON Point: `{ type: 'Point', coordinates: [longitude, latitude] }`) - *Crucial for sellers and customers*
    *   `walletBalance` (Number, default: 0) - *Only relevant for sellers*
    *   `isApproved` (Boolean, default: false) - *For sellers, requires admin approval*
    *   `createdAt`, `updatedAt`
*   **Products:** (Could be simple if only milk, or more complex)
    *   `productId` (Primary Key)
    *   `sellerId` (Foreign Key to Users)
    *   `name` (e.g., \"Cow Milk\", \"Buffalo Milk\")
    *   `description`
    *   `pricePerUnit` (e.g., price per Litre)
    *   `unit` (e.g., \"Litre\", \"500ml\")
    *   `availableStock` (Number)
*   **Orders:**
    *   `orderId` (Primary Key)
    *   `customerId` (Foreign Key to Users)
    *   `sellerId` (Foreign Key to Users)
    *   `items` (Array of `{ productId, quantity, priceAtOrder }`)
    *   `totalAmount`
    *   `status` ('pending', 'accepted', 'rejected', 'delivered', 'cancelled_by_customer', 'cancelled_by_seller')
    *   `orderTimestamp`
    *   `acceptedTimestamp` (null initially)
    *   `cancellationReason` (optional text)
    *   `fineApplied` (Boolean, default: false)
    *   `deliveryAddress` (text)
*   **Transactions:**
    *   `transactionId` (Primary Key)
    *   `userId` (Foreign Key to Users - can be seller or admin)
    *   `type` ('topup', 'order_fee', 'fine_credit', 'fine_debit')
    *   `amount` (Number)
    *   `orderId` (Foreign Key to Orders, optional)
    *   `description` (e.g., \"Fee for order #123\", \"Fine from customer #456\", \"Admin Topup\")
    *   `timestamp`

**2. API Endpoints (Examples):**

*   **Auth:**
    *   `POST /api/auth/register` (customer, seller)
    *   `POST /api/auth/login`
    *   `GET /api/auth/me` (Get logged-in user details)
*   **Users:**
    *   `PUT /api/users/profile` (Update profile, including location)
*   **Customers:**
    *   `GET /api/sellers/nearby?lat=...&lon=...&radius=...` (Find approved sellers nearby)
    *   `GET /api/sellers/:sellerId` (Get seller details and products)
    *   `POST /api/orders` (Place order)
    *   `GET /api/orders/my-orders` (Get customer's order history)
    *   `PUT /api/orders/:orderId/cancel` (Customer cancels order)
        *   *Backend Logic:* Check if order `status` is 'accepted'. If yes, apply ₹12 fine (update order `fineApplied=true`, create 'fine_debit' transaction for customer - *how this is collected needs definition*, create 'fine_credit' transaction for admin). Update order `status` to 'cancelled_by_customer'.
*   **Sellers:**
    *   `GET /api/orders/received` (Get orders for the seller)
    *   `PUT /api/orders/:orderId/accept`
        *   *Backend Logic:* Check seller `walletBalance >= 2`. If yes, deduct ₹2 (update seller `walletBalance`, create 'order_fee' transaction for seller, create corresponding credit transaction for admin). Update order `status` to 'accepted', set `acceptedTimestamp`. If no, return error \"Insufficient balance\".
    *   `PUT /api/orders/:orderId/reject` (Update order `status` to 'rejected')
    *   `PUT /api/orders/:orderId/deliver` (Update order `status` to 'delivered')
    *   `GET /api/wallet/balance`
    *   `GET /api/wallet/transactions`
    *   `POST /api/wallet/request-topup` (Seller requests topup - needs admin action)
    *   `PUT /api/products/:productId` (Update product stock/price)
    *   `POST /api/products` (Add new product)
*   **Super Admin:**
    *   `GET /api/admin/sellers?status=...` (List sellers - pending, approved)
    *   `PUT /api/admin/sellers/:sellerId/approve` (Approve a seller)
    *   `GET /api/admin/transactions` (View all transactions)
    *   `POST /api/admin/sellers/:sellerId/topup` (Admin confirms/adds topup amount to seller wallet, create 'topup' transaction)
    *   `GET /api/admin/overview` (Dashboard stats)

---

**II. Frontend Implementation (React Native)**

**1. Project Setup:**

```bash
# Using Expo
npx create-expo-app MilkDairyApp
cd MilkDairyApp
npm install @react-navigation/native @react-navigation/stack @react-navigation/bottom-tabs react-native-screens react-native-safe-area-context react-native-maps expo-location react-native-elements axios # (or fetch)

# Using React Native CLI
npx react-native init MilkDairyApp --template react-native-template-typescript # (Optional: Use TypeScript)
cd MilkDairyApp
npm install @react-navigation/native @react-navigation/stack @react-navigation/bottom-tabs react-native-screens react-native-safe-area-context react-native-maps react-native-geolocation-service react-native-permissions react-native-elements axios # (or fetch)
# + platform specific setup for maps, location, permissions (Follow library docs)
```

**2. Folder Structure (Suggestion):**

```
MilkDairyApp/
├── src/
│   ├── api/             # API call functions (e.g., auth.js, orders.js)
│   ├── assets/          # Images, fonts
│   ├── components/      # Reusable UI components (Button, Card, Input)
│   ├── constants/       # Colors, dimensions, configuration
│   ├── contexts/        # React Context (e.g., AuthContext.js)
│   ├── hooks/           # Custom hooks (e.g., useLocation.js)
│   ├── navigation/      # Navigation setup (AppNavigator.js, AuthStack.js, etc.)
│   ├── screens/         # Screen components
│   │   ├── Auth/
│   │   │   ├── LoginScreen.js
│   │   │   └── SignupScreen.js
│   │   ├── Customer/
│   │   │   ├── HomeScreen.js       # Map/List of sellers
│   │   │   ├── SellerDetailScreen.js
│   │   │   ├── OrderScreen.js
│   │   │   └── ProfileScreen.js
│   │   ├── Seller/
│   │   │   ├── DashboardScreen.js  # Overview, new orders
│   │   │   ├── OrdersManageScreen.js
│   │   │   ├── WalletScreen.js
│   │   │   └── ProfileManageScreen.js
│   │   └── SuperAdmin/       # Simplified Admin UI in RN
│   │       ├── AdminDashboard.js
│   │       ├── ManageSellersScreen.js
│   │       └── TransactionsScreen.js
│   ├── services/        # Business logic, utility functions
│   └── utils/           # Helper functions
├── App.js               # Entry point
└── ...                  # Other config files (babel, metro, package.json)
```

**3. Core Components & Logic Snippets:**

*   **`src/contexts/AuthContext.js`:**

```javascript
import React, { createContext, useState, useContext, useEffect } from 'react';
import AsyncStorage from '@react-native-async-storage/async-storage';
import api from '../api'; // Your API helper

const AuthContext = createContext(null);

export const AuthProvider = ({ children }) => {
    const [user, setUser] = useState(null);
    const [role, setRole] = useState(null);
    const [isLoading, setIsLoading] = useState(true);

    useEffect(() => {
        // Check async storage for token/user data on app start
        const loadUserData = async () => {
            const storedToken = await AsyncStorage.getItem('userToken');
            if (storedToken) {
                try {
                    // You might want to verify the token with your backend here
                    const userData = await api.auth.getMe(storedToken); // Example API call
                    setUser(userData);
                    setRole(userData.role);
                    // Set token globally for subsequent API calls (in your api helper)
                } catch (error) {
                    console.error(\"Failed to load user data:\", error);
                    await AsyncStorage.removeItem('userToken'); // Clear invalid token
                }
            }
            setIsLoading(false);
        };
        loadUserData();
    }, []);

    const login = async (email, password) => {
        try {
            const { token, userData } = await api.auth.login(email, password);
            await AsyncStorage.setItem('userToken', token);
            setUser(userData);
            setRole(userData.role);
             // Set token globally for subsequent API calls (in your api helper)
        } catch (error) {
            console.error(\"Login failed:\", error);
            throw error; // Re-throw to handle in UI
        }
    };

    const signup = async (userData) => {
        // Similar logic for signup API call
        // ...
    };

    const logout = async () => {
        await AsyncStorage.removeItem('userToken');
        setUser(null);
        setRole(null);
         // Clear global token (in your api helper)
    };

    return (
        <AuthContext.Provider value={{ user, role, login, signup, logout, isLoading }}>
            {children}
        </AuthContext.Provider>
    );
};

export const useAuth = () => useContext(AuthContext);
```

*   **`src/navigation/AppNavigator.js`:**

```javascript
import React from 'react';
import { NavigationContainer } from '@react-navigation/native';
import { createStackNavigator } from '@react-navigation/stack';
import { createBottomTabNavigator } from '@react-navigation/bottom-tabs';
import { useAuth } from '../contexts/AuthContext';

// Import Screens
import LoginScreen from '../screens/Auth/LoginScreen';
import SignupScreen from '../screens/Auth/SignupScreen';
import CustomerHomeScreen from '../screens/Customer/HomeScreen';
import CustomerOrdersScreen from '../screens/Customer/OrderScreen';
import CustomerProfileScreen from '../screens/Customer/ProfileScreen';
import SellerDashboardScreen from '../screens/Seller/DashboardScreen';
import SellerOrdersManageScreen from '../screens/Seller/OrdersManageScreen';
import SellerWalletScreen from '../screens/Seller/WalletScreen';
import SellerProfileManageScreen from '../screens/Seller/ProfileManageScreen';
// Import Admin screens if building RN admin UI
// ... other screens like SellerDetailScreen etc.

const Stack = createStackNavigator();
const CustomerTab = createBottomTabNavigator();
const SellerTab = createBottomTabNavigator();
// const AdminTab = createBottomTabNavigator(); // If applicable

function CustomerNavigator() {
    return (
        <CustomerTab.Navigator>
            <CustomerTab.Screen name=\"Home\" component={CustomerHomeScreen} options={{ title: 'Nearby Sellers' }} />
            <CustomerTab.Screen name=\"Orders\" component={CustomerOrdersScreen} options={{ title: 'My Orders' }}/>
            <CustomerTab.Screen name=\"Profile\" component={CustomerProfileScreen} />
             {/* Add other tabs/screens like Cart if needed */}
        </CustomerTab.Navigator>
    );
}

function SellerNavigator() {
    return (
        <SellerTab.Navigator>
            <SellerTab.Screen name=\"Dashboard\" component={SellerDashboardScreen} />
            <SellerTab.Screen name=\"Manage Orders\" component={SellerOrdersManageScreen} />
            <SellerTab.Screen name=\"Wallet\" component={SellerWalletScreen} />
            <SellerTab.Screen name=\"Shop Profile\" component={SellerProfileManageScreen} />
        </SellerTab.Navigator>
    );
}

// Add AdminNavigator similarly if needed

function AuthNavigator() {
    return (
        <Stack.Navigator screenOptions={{ headerShown: false }}>
            <Stack.Screen name=\"Login\" component={LoginScreen} />
            <Stack.Screen name=\"Signup\" component={SignupScreen} />
        </Stack.Navigator>
    );
}

export default function AppNavigator() {
    const { user, role, isLoading } = useAuth();

    if (isLoading) {
        // Optionally return a Loading Spinner screen
        return null; // Or a loading indicator
    }

    return (
        <NavigationContainer>
            {!user ? (
                <AuthNavigator />
            ) : role === 'customer' ? (
                 // You might want a stack navigator wrapping the tabs
                 // to handle screens like Seller Details which are not tabs
                 <Stack.Navigator>
                      <Stack.Screen name=\"CustomerMain\" component={CustomerNavigator} options={{ headerShown: false }}/>
                      {/* Add SellerDetailScreen, OrderConfirmationScreen etc. here */}
                 </Stack.Navigator>
            ) : role === 'seller' ? (
                 <Stack.Navigator>
                    <Stack.Screen name=\"SellerMain\" component={SellerNavigator} options={{ headerShown: false }}/>
                     {/* Add OrderDetailScreen for seller here */}
                 </Stack.Navigator>
            ) : role === 'superadmin' ? (
                // <AdminNavigator /> // If implemented
                <Text>Admin Panel Placeholder</Text> // Placeholder
            ) : (
                 // Fallback to Auth if role is unknown or user exists without role
                 <AuthNavigator />
            )}
        </NavigationContainer>
    );
}
```

*   **`src/screens/Customer/HomeScreen.js` (Conceptual):**

```javascript
import React, { useState, useEffect } from 'react';
import { View, Text, StyleSheet, FlatList, ActivityIndicator, PermissionsAndroid, Platform } from 'react-native';
import MapView, { Marker } from 'react-native-maps';
import * as Location from 'expo-location'; // Or use react-native-geolocation-service
import api from '../../api'; // Your API helper
import SellerCard from '../../components/SellerCard'; // Create this component
import { useNavigation } from '@react-navigation/native';

const HomeScreen = () => {
    const [location, setLocation] = useState(null);
    const [sellers, setSellers] = useState([]);
    const [loading, setLoading] = useState(true);
    const [errorMsg, setErrorMsg] = useState(null);
    const navigation = useNavigation();

    useEffect(() => {
        (async () => {
            let { status } = await Location.requestForegroundPermissionsAsync(); // Expo example
            // For RN CLI, use PermissionsAndroid or react-native-permissions
            if (status !== 'granted') {
                setErrorMsg('Permission to access location was denied');
                setLoading(false);
                return;
            }

            try {
                let currentLocation = await Location.getCurrentPositionAsync({});
                setLocation(currentLocation.coords);
                fetchNearbySellers(currentLocation.coords);
            } catch (error) {
                 setErrorMsg('Could not get location. Please enable location services.');
                 setLoading(false);
                 console.error(error);
            }
        })();
    }, []);

    const fetchNearbySellers = async (coords) => {
        setLoading(true);
        try {
            // Adjust radius as needed (e.g., 5km = 5000m)
            const response = await api.sellers.findNearby(coords.latitude, coords.longitude, 5000);
            setSellers(response.data); // Assuming API returns an array of sellers
            setErrorMsg(null);
        } catch (error) {
            console.error(\"Error fetching sellers:\", error);
            setErrorMsg('Failed to load nearby sellers.');
            setSellers([]);
        } finally {
            setLoading(false);
        }
    };

    const handleSellerPress = (seller) => {
        navigation.navigate('SellerDetail', { sellerId: seller.userId }); // Navigate to seller detail
    };

    return (
        <View style={styles.container}>
            {loading && <ActivityIndicator size=\"large\" style={styles.loader} />}
            {errorMsg && <Text style={styles.errorText}>{errorMsg}</Text>}

            {location && (
                <MapView
                    style={styles.map}
                    initialRegion={{
                        latitude: location.latitude,
                        longitude: location.longitude,
                        latitudeDelta: 0.0922,
                        longitudeDelta: 0.0421,
                    }}
                    showsUserLocation={true}
                >
                    {sellers.map(seller => (
                        <Marker
                            key={seller.userId}
                            coordinate={{
                                latitude: seller.location.coordinates[1], // Assuming [lon, lat] format
                                longitude: seller.location.coordinates[0],
                            }}
                            title={seller.shopName || seller.name}
                            description=\"Tap to view products\"
                            onCalloutPress={() => handleSellerPress(seller)} // Open detail on bubble press
                        />
                    ))}
                </MapView>
            )}

            <Text style={styles.listHeader}>Nearby Sellers:</Text>
            {sellers.length > 0 ? (
                <FlatList
                    data={sellers}
                    keyExtractor={(item) => item.userId}
                    renderItem={({ item }) => (
                         <SellerCard seller={item} onPress={() => handleSellerPress(item)} />
                    )}
                    style={styles.list}
                />
            ) : (
                 !loading && <Text style={styles.noSellersText}>No sellers found nearby.</Text>
            )}
        </View>
    );
};

const styles = StyleSheet.create({
    container: { flex: 1 },
    map: { height: '50%', width: '100%' },
    loader: { position: 'absolute', top: '50%', left: '50%', marginLeft: -25, marginTop: -25 },
    errorText: { color: 'red', textAlign: 'center', padding: 10 },
    listHeader: { fontSize: 18, fontWeight: 'bold', padding: 10, backgroundColor: '#eee' },
    list: { flex: 1 },
    noSellersText: { textAlign: 'center', marginTop: 20, fontSize: 16, color: 'grey' },
    // ... Add styles for SellerCard
});

export default HomeScreen;
```

*   **`src/screens/Seller/WalletScreen.js` (Conceptual):**

```javascript
import React, { useState, useEffect, useCallback } from 'react';
import { View, Text, StyleSheet, Button, FlatList, ActivityIndicator, RefreshControl } from 'react-native';
import api from '../../api'; // Your API helper
import { useFocusEffect } from '@react-navigation/native'; // To refresh on screen focus

const WalletScreen = () => {
    const [balance, setBalance] = useState(null);
    const [transactions, setTransactions] = useState([]);
    const [loading, setLoading] = useState(true);
    const [refreshing, setRefreshing] = useState(false);
    const [error, setError] = useState(null);

    const fetchData = async () => {
        setLoading(true);
        setError(null);
        try {
            const [balanceRes, transactionsRes] = await Promise.all([
                api.wallet.getBalance(),
                api.wallet.getTransactions()
            ]);
            setBalance(balanceRes.data.balance); // Adjust based on your API response structure
            setTransactions(transactionsRes.data); // Adjust based on your API response structure
        } catch (err) {
            console.error(\"Error fetching wallet data:\", err);
            setError(\"Failed to load wallet details. Pull down to retry.\");
        } finally {
            setLoading(false);
            setRefreshing(false);
        }
    };

    // Fetch data when the screen comes into focus
     useFocusEffect(
        useCallback(() => {
            fetchData();
        }, [])
    );

    const onRefresh = useCallback(() => {
        setRefreshing(true);
        fetchData();
    }, []);

    const handleRequestTopup = async () => {
        // Implement UI for amount input if needed, then call API
        try {
            await api.wallet.requestTopup({ amount: 500 }); // Example fixed amount
            alert('Top-up request sent. Admin will process it soon.');
            // Maybe refresh data or update UI optimistically
        } catch (error) {
            console.error(\"Topup request failed:\", error);
            alert('Failed to send top-up request.');
        }
    };

    const renderTransaction = ({ item }) => (
        <View style={styles.transactionItem}>
            <Text style={styles.transDescription}>{item.description || `Type: ${item.type}`}</Text>
            <Text style={[styles.transAmount, { color: item.amount > 0 ? 'green' : 'red' }]}>
                 ₹{item.amount.toFixed(2)}
            </Text>
             <Text style={styles.transDate}>{new Date(item.timestamp).toLocaleString()}</Text>
        </View>
    );

    return (
        <View style={styles.container}>
            <View style={styles.balanceContainer}>
                <Text style={styles.balanceLabel}>Current Balance:</Text>
                {loading && balance === null ? (
                     <ActivityIndicator color=\"#007AFF\"/>
                ) : error && balance === null ? (
                     <Text style={styles.errorText}>Error loading balance</Text>
                 ): (
                    <Text style={styles.balanceAmount}>₹{balance !== null ? balance.toFixed(2) : 'N/A'}</Text>
                )}
                <Button title=\"Request Top-up (Needs Admin Approval)\" onPress={handleRequestTopup} />
            </View>

            <Text style={styles.listHeader}>Transaction History</Text>
            {loading && transactions.length === 0 ? (
                 <ActivityIndicator size=\"large\" style={{marginTop: 20}} />
             ) : error ? (
                 <Text style={[styles.errorText, {textAlign: 'center', marginTop: 20}]}>{error}</Text>
             ) : (
                 <FlatList
                    data={transactions}
                    keyExtractor={(item) => item.transactionId}
                    renderItem={renderTransaction}
                    ListEmptyComponent={<Text style={styles.emptyText}>No transactions yet.</Text>}
                    refreshControl={
                        <RefreshControl refreshing={refreshing} onRefresh={onRefresh} />
                    }
                    style={styles.list}
                />
             )}
        </View>
    );
};

const styles = StyleSheet.create({
    container: { flex: 1, backgroundColor: '#f8f8f8' },
    balanceContainer: { padding: 20, backgroundColor: 'white', marginBottom: 10, alignItems: 'center' },
    balanceLabel: { fontSize: 16, color: 'grey' },
    balanceAmount: { fontSize: 32, fontWeight: 'bold', marginVertical: 10 },
    listHeader: { fontSize: 18, fontWeight: 'bold', padding: 10, backgroundColor: '#eee' },
    list: { flex: 1 },
    transactionItem: {
        backgroundColor: 'white',
        padding: 15,
        borderBottomWidth: 1,
        borderBottomColor: '#eee',
        flexDirection: 'row', // Arrange items horizontally
        justifyContent: 'space-between', // Space out elements
        alignItems: 'center' // Align vertically
    },
    transDescription: { flex: 2, fontSize: 15 }, // Allow description to take more space
    transAmount: { flex: 1, fontSize: 16, fontWeight: 'bold', textAlign: 'right' },
    transDate: { flex: 1.5, fontSize: 12, color: 'grey', textAlign: 'right', marginLeft: 5 }, // Date on the right
    emptyText: { textAlign: 'center', marginTop: 30, color: 'grey' },
    errorText: { color: 'red', fontSize: 14 },
    // ... other styles
});

export default WalletScreen;
```

*   **Order Cancellation Logic (Customer Side - Conceptual):**

```javascript
// In CustomerOrdersScreen or OrderDetailScreen

const handleCancelOrder = async (order) => {
    // Check if cancellation is allowed (e.g., status is 'pending' or 'accepted')
    if (order.status !== 'pending' && order.status !== 'accepted') {
        alert(\"Cannot cancel an order that is already \" + order.status);
        return;
    }

    const fineWarning = order.status === 'accepted'
        ? \"This order has been accepted by the seller. Cancelling now will incur a ₹12 fine. Are you sure?\"
        : \"Are you sure you want to cancel this order?\";

    Alert.alert(
        \"Confirm Cancellation\",
        fineWarning,
        [
            { text: \"No\", style: \"cancel\" },
            {
                text: \"Yes, Cancel\",
                onPress: async () => {
                    try {
                        setCancelling(true); // Show loading indicator
                        await api.orders.cancelOrder(order.orderId); // Call backend API
                        alert(\"Order Cancelled Successfully.\");
                        // Refresh the order list
                        fetchOrders();
                    } catch (error) {
                        console.error(\"Cancellation failed:\", error);
                        alert(\"Failed to cancel order. \" + (error.response?.data?.message || \"Please try again.\"));
                    } finally {
                         setCancelling(false);
                    }
                },
                style: \"destructive\" // Red button for destructive action
            }
        ]
    );
};

// Add a button in your order list item or detail screen:
<Button title=\"Cancel Order\" onPress={() => handleCancelOrder(order)} color=\"red\" disabled={cancelling} />
```

*   **Order Acceptance Logic (Seller Side - Conceptual):**

```javascript
// In SellerOrdersManageScreen or OrderDetailScreen

const handleAcceptOrder = async (order) => {
    if (order.status !== 'pending') {
        alert(\"Order is not pending.\");
        return;
    }

    Alert.alert(
        \"Confirm Acceptance\",
        \"Accepting this order will deduct ₹2 from your wallet. Proceed?\",
        [
            { text: \"Cancel\", style: \"cancel\" },
            {
                text: \"Accept Order\",
                onPress: async () => {
                    try {
                        setAccepting(true); // Show loading indicator
                        await api.orders.acceptOrder(order.orderId);
                        alert(\"Order Accepted. ₹2 deducted from wallet.\");
                        // Refresh order list or update order status locally
                        fetchReceivedOrders();
                    } catch (error) {
                        console.error(\"Acceptance failed:\", error);
                         // Check for specific insufficient balance error from backend
                         if (error.response?.data?.code === 'INSUFFICIENT_BALANCE') {
                             alert(\"Failed to accept order: Insufficient wallet balance. Please top up.\");
                         } else {
                             alert(\"Failed to accept order. \" + (error.response?.data?.message || \"Please try again.\"));
                         }
                    } finally {
                         setAccepting(false);
                    }
                }
            }
        ]
    );
};

// Add a button:
<Button title=\"Accept Order\" onPress={() => handleAcceptOrder(order)} color=\"green\" disabled={accepting} />
```

**4. UI/UX Improvements:**

*   **Loading States:** Use `ActivityIndicator` extensively during data fetching, location checks, and button actions.
*   **Error Handling:** Display user-friendly error messages (e.g., using banners, toasts, or inline text). Avoid raw error logs.
*   **Feedback:** Provide clear feedback on actions (e.g., \"Order placed successfully\", \"Wallet topped up\"). Use toast messages (`react-native-toast-message`).
*   **Empty States:** Design informative screens for when lists are empty (no orders, no sellers found, no transactions).
*   **Forms:** Use form libraries (`react-hook-form`) for validation and better state management in Login/Signup/Profile forms.
*   **Map Interaction:** Allow users to tap on map markers to see seller info and navigate. Consider clustering markers if many sellers are close.
*   **Clear Navigation:** Use intuitive icons and labels for bottom tabs. Ensure easy back navigation within stacks.
*   **Consistent Styling:** Define a color palette, typography, and component styles in `src/constants` or use a theme from UI libraries.
*   **Permissions:** Handle location permission requests gracefully. Explain *why* the permission is needed before asking. Provide a way for the user to grant permission later if they initially deny it.
*   **Pull-to-Refresh:** Implement pull-to-refresh on lists (Orders, Transactions, Seller List) for easy data updates.
*   **Visual Hierarchy:** Use font sizes, weights, and colors effectively to guide the user's eye. Use components like `Card` for grouping related information.

---

**III. Important Considerations:**

1.  **Backend Implementation:** This is crucial. The frontend code relies entirely on the backend API being available and functional. Choose a technology you're comfortable with (Node.js, Python, etc.) and a suitable database.
2.  **Real-time Updates:** For order status changes, consider using WebSockets or Server-Sent Events (SSE) for a better UX, rather than relying solely on manual refresh.
3.  **Payment Gateway:** For seller top-ups, you'll need to integrate a payment gateway (like Stripe, Razorpay, Paytm). This involves server-side handling and secure token exchange. The conceptual `request-topup` and admin confirmation is a simplified alternative.
4.  **Fine Collection:** Define *how* the customer fine is actually collected. Is it added to their next order? Do they need to pay it separately? This impacts backend logic and potentially UI.
5.  **Security:** Implement proper authentication (JWT recommended), authorization (ensure users can only access their own data or data relevant to their role), input validation, and HTTPS.
6.  **Scalability:** Design your backend and database schema with potential growth in mind.
7.  **Admin Panel:** A web-based admin panel is generally more practical for managing users, approvals, transactions, and viewing comprehensive reports than a React Native one.
8.  **Testing:** Thoroughly test all user flows, edge cases (low balance, cancelled orders, permission denial), and API interactions.

This detailed outline and conceptual code should give you a strong foundation for building your Milk Dairy app. Remember to replace the conceptual API calls with actual `fetch` or `axios` requests to your real backend endpoints once built."""),
            ],
        ),
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(text="""INSERT_INPUT_HERE"""),
            ],
        ),
    ]
    generate_content_config = types.GenerateContentConfig(
        response_mime_type="text/plain",
        system_instruction=[
            types.Part.from_text(text="""Preview
"""),
        ],
    )

    for chunk in client.models.generate_content_stream(
        model=model,
        contents=contents,
        config=generate_content_config,
    ):
        print(chunk.text, end="")

if __name__ == "__main__":
    generate()
````
